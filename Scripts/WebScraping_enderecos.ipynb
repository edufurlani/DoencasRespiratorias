```python
# WebScraping

enderecos_encontrados = []
enderecos_nao_encontrados = []

def obter_endereco_hospital(hospital):
    pesquisa = f"Endereço do hospital {hospital}"
    url_pesquisa = f"https://www.google.com.br/search?q={pesquisa}"

    # Faz a solicitação HTTP para a pesquisa no Google
    resposta = requests.get(url_pesquisa)

    # Analisa a página de resultados do Google
    if resposta.status_code == 200:
        soup = BeautifulSoup(resposta.text, 'html.parser')

        # Encontra o elemento que contém o endereço (pode variar dependendo do site)
        endereco_element = soup.find('div', class_='BNeawe iBp4i AP7Wnd')

        if endereco_element:
            endereco = endereco_element.text
            enderecos_encontrados.append({'Hospital': hospital, 'Endereço': endereco})
        else:
            enderecos_nao_encontrados.append(hospital)

# Iterar sobre cada hospital e obter o endereço
for hospital in hospitais:
    obter_endereco_hospital(hospital)

# Criar DataFrames a partir das listas de resultados
df_enderecos_encontrados = pd.DataFrame(enderecos_encontrados)
df_enderecos_nao_encontrados = pd.DataFrame({'Hospital': enderecos_nao_encontrados})

# Salvar os resultados em arquivos Excel separados
df_enderecos_encontrados.to_excel("enderecos_hospitais_encontrados.xlsx", index=False)
#df_enderecos_nao_encontrados.to_excel("enderecos_hospitais_nao_encontrados.xlsx", index=False)

print("Endereços encontrados salvos no arquivo 'enderecos_hospitais_encontrados.xlsx'")
print("Hospitais sem endereço encontrado salvos no arquivo 'enderecos_hospitais_nao_encontrados.xlsx'")
```
